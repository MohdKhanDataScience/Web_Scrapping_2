```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
description=[]
processor=[]
ram=[]
os=[]
storage=[]
screen_size=[]
warrenty=[]
price=[]
```

```python
pages=list(range(1,24))
for _ in pages:
    req=requests.get('https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
    soup=BeautifulSoup(req.text,'html.parser')
    desc=soup.find_all(class_='_4rR01T')
    for i in range(len(desc)):
        description.append(desc[i].text)
    common_class=soup.find_all(class_='rgWa7D')
    for i in range(len(common_class)):
        s=common_class[i].text
        if('Core' in s):
            processor.append(s)
        elif('RAM' in s):
            ram.append(s)
        elif('Operating' in s):
            os.append(s)
        elif(('SSD' or'HDD') in s):
            storage.append(s)
        elif('Display' in s):
            screen_size.append(s)
        elif('Warranty' in s):
            warrenty.append(s)
    price_class=soup.find_all(class_='_30jeq3 _1_WHN1')
    for i in range(len(price_class)):
        price.append(price_class[i].text)
```

```python
print(len(description))
print(len(processor))
print(len(ram))
print(len(storage))
print(len(os))
print(len(screen_size))
print(len(warrenty))
print(len(price))
```

```python
d={'Description':description[:379],'Processor':processor[:379],'RAM':ram[:379],'OperatingSystem':os[:379],
  'Storage':storage[:379],'Screen_size':screen_size[:379],'Warranty':warrenty[:379],'Price':price[:379]}
```

```python
df=pd.DataFrame(d)
```
